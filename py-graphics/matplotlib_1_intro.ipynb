### plot( )로 그래프를 그린다

```python
np.arange(0.0, 2.0, 0.01) # np.arange(시작 값, 끝 값, 단계) -> 0.0부터 2.0까지 0.01씩 증가하는 값
```

```python
len(np.arange(0.0, 2.0, 0.01))  # 길이는 200개
```

```python
np.arange(0.0, 2.0, 0.01).shape # 200개 요소로 구성된 1차원 벡터라는 것을 알려주는 튜플
```

```python
# sin그래프를 그린다
s = np.sin(np.pi*np.arange(0.0, 2.0, 0.01))
plt.plot(s)    # sin의 그래프를 그림  # x축이 data의 인덱스로 돼있음(200)
```

### x-축을 지정하여 그리기.

```python
s1 = np.sin(np.pi*np.arange(0.0, 2.0, 0.1))
plt.plot(np.arange(0.0, 2.0, 0.1), s1, '*' , ms = 10) # plt.plot(x축, y축, mark 모양, mark 크기)
```

## 산포도 (scatter plot)

```python
# 산포도(scatter plot)를 그린다
# 마커를 사용하여 산포도를 작성
# 표준 정규 분포에 따라 난수 생성
x = np.random.randn(5000) # 가우시안
y = np.random.randn(5000) # 가우시안
plt.figure(figsize=(5,5)) # figsize - 내가 그리는 그림의 사이즈를 지정
t = plt.plot(x, y, 'o', alpha=0.1)   #, ms=8) # alpha - 불투명도
```

### 막대 그래프 (Bar graph)

```python
# 막대그래프(bar graph)를 그린다.
means = [1, 2, 3] # 평균이 1, 2, 3
stddevs = [0.2, 0.4, 0.5] # 평균 1의 표준편차 0.2, 평균 2의 표준편차 0.4, 평균 3의 표준편차 0.5
bar_labels = ['bar 1', 'bar 2', 'bar 3']

# plot bars
x_pos = list(range(len(bar_labels)))  # 각 그래프들의 위치를 정해줌.
                                      # 첫번째 bar - 1, 두번째 bar - 2, 세번째 bar - 3
plt.bar(x_pos, means, yerr=stddevs, color=('r','g','b'))

plt.show()
```

### 히스토그램 (histogram)

```python
# 히스토그램 (histogram)
rng = np.random.RandomState(123)  # different from random.seed()
x = rng.normal(0, 20, 1000) # RandomState로 만든 무작위 수 정규분포 평균 0, 표준편자 20인 것 1000개

# fixed bin size
bins = np.arange(-100, 100, 5) # fixed bin size -> -100부터 100까지 5단위로 약 40개의 막대

plt.hist(x, bins=bins)
plt.show()
```

### [도전하기] 위의 histogram을 정수 bins를 지정해서 그리시오.






```python
# 히스토그램 (histogram)
rng = np.random.RandomState(123)  # different from random.seed()
x = rng.normal(0, 20, 1000) 

# fixed bin size
# bins = np.arange(-100, 100, 5) # fixed bin size 
bins = 100
 
plt.hist(x, bins=bins)
plt.show()
```

```python
# 중첩 히스토그램
rng = np.random.RandomState(123)
x1 = rng.normal(0, 20, 1000) 
x2 = rng.normal(20, 10, 1000) # 평균이 20, 표준편차 10 가우시안 1000개
```

```python
x1.shape, x2.shape
```

```python
# fixed bin size
bins = np.arange(-100, 100, 5) # fixed bin size

plt.hist(x1, bins=bins, alpha=0.5)
plt.hist(x2, bins=bins, alpha=0.5)
plt.show()
```
